```python

def plot_some_lines():
    plt.figure()
    x1 = [10,20,5,40,8]
    x2 = [30,43,9,7,20]
    plt.plot(x1, label="Group Spam")
    plt.plot(x2, label="Group Eggs")
    plt.legend()
    plt.show()

plot_some_lines()
```

```python
car_crash= sns.load_dataset("car_crashes")
car_crash
```

# Show table of data

```python
car_crash.head(5)
```

```python
sns.relplot(x="alcohol", y="speeding", hue="speeding", data=car_crash)
```

# As one consumes more alcohol, the rate of speed increases.

```python
states = car_crash.groupby("alcohol", as_index=False).sum()
sns.relplot(x="alcohol", y="total", data=states)
sns.lmplot(x="alcohol", y="total", data=states)
```

# Conclusions
- There is an upward trend of when alcohol increases, so does speeding. This may lead to an increased rate of crashes as alcohol consumption is raised. The more alcohol consumed leads to a greater total of crashes.

# Visual Two

```python
p = sns.load_dataset("penguins")
p
```

```python
p.head(5)
```

```python
sns.relplot(x="body_mass_g", y="flipper_length_mm", hue="island", data=p)
```

```python
bill = p.groupby("bill_length_mm", as_index=False).sum()
sns.relplot(x='bill_length_mm', y="body_mass_g", data=bill)
```

# Conclusions
- As body mass increases, so does the flipper length of the penguins. Penugins from the Biscoe island are of bigger nature compared to those from Dream or Torgersen island. It can also be assumed that the bill length of penguins seems to increase as body mass increases up until around a bill length of 45mm, and then it drops back down. This is not considered to be a fully positive correlation between body mass and bill length.
